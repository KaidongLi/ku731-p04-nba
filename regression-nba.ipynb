At the first glance, we notice there are many NaN values in the data. So we want a count of how many NaN rows each of the columns has. And we want to develop a strategy to handle it.

```python
coln_null = d_ori.columns[d_ori.isnull().any()]
d_ori[coln_null].isnull().sum()
```

```python
data_pro = d_ori.drop(columns=['playoff', 'carm-elo1_pre', 'carm-elo2_pre', 'carm-elo1_post', 'carm-elo2_post', 'carm-elo_prob1', 'carm-elo_prob2', 'carmelo1_pre', 'carmelo2_pre', 'carmelo_prob1', 'carmelo_prob2'])
data_pro = data_pro.dropna()
data_pro.sample(2)
```

```python
# encoding from string to number
from sklearn import preprocessing
ftr_team1 = ['team1']
ftr = ['team1', 'team2']

le = preprocessing.LabelEncoder()
le = le.fit(data_pro['team1'])
data_pro['team1'] = le.transform(data_pro['team1'])
data_pro['team2'] = le.transform(data_pro['team2'])

#data_pro.sample(3)
```

Seasons are important to a team. They will have different level of skills because of recruitments. So we will calculate the win-loss ratio of each season. And the month-date information is not that helpful. The performance of each team is determined by some random factors. So we will get rid of that.

```python
del data_pro['date']
```

```python
data_pro.sample(2)
```

Creat a dataframe to store winning/loss record of each team and season

```python
df_wl = pd.DataFrame(columns=['season', 'team', 'win', 'loss', 'tie', 'winning_rate'])
l_sn = data_pro['season'].unique()
l_team = data_pro['team1'].unique()
idx = 0
for season in l_sn:
    for team in l_team:
        df_temp = pd.DataFrame({'season': season, 'team': team, 'win': 0, 'loss': 0, 'tie': 0, 'winning_rate': -1}, index=[idx])
        df_wl = df_wl.append(df_temp)
        idx += 1
```

```python
for i, row in data_pro.iterrows():
    if (row['score1'] > row['score2']):
        l_temp = df_wl[ (df_wl.season == row.season) & (df_wl.team == row.team1) ]
        df_wl.loc[l_temp.index, 'win'] += 1
        l_temp = df_wl[ (df_wl.season == row.season) & (df_wl.team == row.team2) ]
        df_wl.loc[l_temp.index, 'loss'] += 1
    elif (row['score1'] < row['score2']):
        l_temp = df_wl[ (df_wl.season == row.season) & (df_wl.team == row.team1) ]
        df_wl.loc[l_temp.index, 'loss'] += 1
        l_temp = df_wl[ (df_wl.season == row.season) & (df_wl.team == row.team2) ]
        df_wl.loc[l_temp.index, 'win'] += 1
    else:
        l_temp = df_wl[ (df_wl.season == row.season) & (df_wl.team == row.team1) ]
        df_wl.loc[l_temp.index, 'tie'] += 1
        l_temp = df_wl[ (df_wl.season == row.season) & (df_wl.team == row.team2) ]
        df_wl.loc[l_temp.index, 'tie'] += 1
```

```python
df_wl.sample(2)
```

Calculate winning ratio

```python
for i, row in df_wl.iterrows():
    c_match = row.loss + row.tie + row.win
    ratio = 0
    if (c_match > 0):
        ratio = float(row.win) / c_match
    df_wl.loc[i, 'winning_rate'] = ratio
```

# Regression
we choose random forest classifier to do regression. We use all the other parameters left in DataFrame as the input. And we try to train on 80% of the data. And using the rest 20% as testing.

```python
from sklearn.model_selection import train_test_split

data_x = data_pro.drop(columns=['score1', 'score2'])
data_y = data_pro[['score1', 'score2']]

perc_test = 0.20
X_train, X_test, y_train, y_test = train_test_split(data_x, data_y, test_size=perc_test)

#data_x
#data_y
#X_train
#y_test
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import make_scorer, accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.multioutput import MultiOutputClassifier

clfr = RandomForestClassifier()
multi_clfr = MultiOutputClassifier(clfr, n_jobs=-1)

#acc_scorer = make_scorer(accuracy_score)

multi_clfr.fit(X_train, y_train)
```

```python
#acc_scorer = make_scorer(accuracy_score)
predictions = multi_clfr.predict(X_test)
#print(accuracy_score(y_test[], predictions))
```

Since accuracy_score doesn't support multiclass multioutput, so we will compare colomn by column

```python
print 'accuracy of team 1'
print(accuracy_score(y_test.score1, predictions[:, 0]))
```

```python
print 'accuracy of team 1'
print(accuracy_score(y_test.score2, predictions[:, 1]))
```
